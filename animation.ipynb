# Save animated GIF of GRIMP DATA


```python
import stackstac
import dask.array as da
import grimpfunc as grimp
import nisardev as nisar
from geogif import gif
```

```python
# Load Xarray datacube over Jacobshavn
# NASA CMR STAC kind of broken
# Instead of different assets, each image type is a separate item, also no PROJ extension...
#https://cmr.earthdata.nasa.gov/stac/NSIDC_ECS/collections/NSIDC-0723.v4/2023/01/01
# https://cmr.earthdata.nasa.gov/stac/NSIDC_ECS/collections/NSIDC-0723.v4/items/SC:NSIDC-0723.004:264648391
# https://cmr.earthdata.nasa.gov/stac/NSIDC_ECS/collections/NSIDC-0723.v4/items/SC:NSIDC-0723.004:264648396
```

```python
urls = grimp.cmrUrls(mode='image') 

# SUBSET to specific glacier and year for testing
bbox = {'minx': 440000, 'miny': -1140000, 'maxx': 500000, 'maxy': -1080000}
urls.initialSearch(firstDate='2020-01-01', lastDate='2021-01-01', productFilter='gamma0');
```

```python
# NOTE: should be faster... 18s initially
myImageSeries = nisar.nisarImageSeries()  # Instantiate the series object
myImageSeries.readSeriesFromTiff(urls.getCogs(), chunkSize=2048) 
myImageSeries.subSetImage(bbox)
myImageSeries.subset
```

```python
# Speed this up / ensure multiple cores! 
myImageSeries.loadRemote()
```
